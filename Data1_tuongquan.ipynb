```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# c d liu t file CSV
data = pd.read_csv('Data1.csv')

# Lm sch d liu
for column in data.columns:
    if data[column].dtype == object:
        data[column] = data[column].replace({',': '.'}, regex=True)
    data[column] = pd.to_numeric(data[column], errors='coerce')
data = data.dropna()

# Ly tt c cc ct (16 thuc tnh)
all_features = data.columns

# Bc 1: V biu  phn b (Histogram) khng c ng cong mt 
plt.figure(figsize=(20, 15))
for i, column in enumerate(all_features, 1):
    plt.subplot(4, 4, i)
    sns.histplot(data[column], bins=20, kde=False)  # Loi b kde bng cch t kde=False
    plt.title(column)
    plt.xlabel('')
    plt.ylabel('')
plt.tight_layout()
plt.savefig('data1_all_histograms_no_kde.png')
plt.show()

```
