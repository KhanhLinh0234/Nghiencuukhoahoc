```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# c d liu t file CSV
data = pd.read_csv('Data1.csv')

# Bc 1: Lm sch d liu
# X l gi tr c_sg c du phy (v d: "2,161" -> 2.161)
data['c_sg'] = data['c_sg'].replace({',': '.'}, regex=True).astype(float)

# Kim tra v loi b gi tr thiu
data = data.dropna()

# Bc 2: Tnh h s tng quan Pearson
correlation_matrix = data.corr(method='pearson')

# Bc 3: Trc quan ha ma trn tng quan
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, fmt='.2f', linewidths=0.5)
plt.title('Dataset 1 Correlation Matrix (Pearson)', fontsize=16)
plt.tight_layout()

# Lu biu 
plt.savefig('correlation_matrix.png')
plt.show()

# In ma trn tng quan ra console
print("\nMa trn tng quan Pearson:")
print(correlation_matrix)
```
