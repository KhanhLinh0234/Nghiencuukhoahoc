```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
data = pd.read_csv('Data1.csv')

# Bước 1: Làm sạch dữ liệu
# Xử lý giá trị c_sg có dấu phẩy (ví dụ: "2,161" -> 2.161)
data['c_sg'] = data['c_sg'].replace({',': '.'}, regex=True).astype(float)

# Kiểm tra và loại bỏ giá trị thiếu
data = data.dropna()

# Bước 2: Tính hệ số tương quan Pearson
correlation_matrix = data.corr(method='pearson')

# Bước 3: Trực quan hóa ma trận tương quan
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, fmt='.2f', linewidths=0.5)
plt.title('Dataset 1 Correlation Matrix (Pearson)', fontsize=16)
plt.tight_layout()

# Lưu biểu đồ
plt.savefig('correlation_matrix.png')
plt.show()

# In ma trận tương quan ra console
print("\nMa trận tương quan Pearson:")
print(correlation_matrix)
```
